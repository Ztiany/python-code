```python
# 导入 urllib3 模块的所有类与对象【Python 以系统类的形式提供了下载网页的功能，放在 urllib3 这个模块中。】
import urllib3

# 将要下载的网址保存在 url 变量中，英文一般用 url 表示网址的意思
jiandan_url = "http://jandan.net/p/date/2021/03/23"

# 创建一个 PoolManager 对象，命名为 http
http = urllib3.PoolManager()

# 调用 http 对象的 request 方法，第一个参数传一个字符串 "GET"
# 第二个参数则是要下载的网址，也就是我们的 url 变量
# request 方法会返回一个 HTTPResponse 类的对象，我们命名为 response
jiandan_response = http.request("GET", jiandan_url)

# 获取 response 对象的 data 属性，存储在变量 response_data 中
jiandan_data = jiandan_response.data

# 调用 response_data 对象的 decode 方法，获得网页的内容，存储在 html_content 变量中
jiandan_content = jiandan_data.decode()

# 打印 html_content
print(jiandan_content)

# 保存到本地文件
# 调用 open 函数，三个参数都是字符串类型：
#   第一个参数为要操作的文件名
#   第二个参数代表模式，w 表示写入文件，r 表示读取文件
#   第三个参数表示编码格式，一般有中文的认准 utf-8 就好
# open 函数返回一个文件对象，我们存储在 fo 变量中
jiandan_file = open("jiandan.html", "w", encoding="utf-8")

# 调用文件对象的 write 方法，将我们上面存储着网页内容的字符春变量，html_content
# 作为参数
jiandan_file.write(jiandan_content)

# 关闭文件对象
jiandan_file.close()
```

代码说明：这是通过 urllib3 请求一个网页的示例。下面是一些补充说明：

1. 我们创建 PoolManager的时候，写的是 urllib3.PoolManager，这里是因为我们导入了 urllib3 的所有类与函数。所以在调用这个模块的所有函数和类的前面都需要加模块名，并用点符号连接。
2. response 对象的 data 属性也是一个对象，是一个 bytes 类型的对象。通过调用 decode 方法，可以转化成我们熟悉的字符串。

```python
import urllib3

def download_content(target_url):
    http_manager = urllib3.PoolManager()
    http_response = http_manager.request("GET", target_url)
    data = http_response.data
    return data.decode()

def save_to_file(filename, file_content):
    file = open(filename,"w", encoding="utf-8")
    file.write(file_content)
    file.close()
```

上面是将下载逻辑封装为函数。下面编码使用上面的函数下载煎蛋的最近 5 页内容：

```python
# i 从 1 到 5 循环。
for i in range(1, 6):
    url = "http://jiandan.net"
    # 第一页不需要加/page/x 后缀，所以判断大于 1 才加
    if i > 1:
        url = url + "/page/" + str(i)
    content = download_content(url)
    save_to_file("jiandan_"+str(i) + ".html", content)
```
